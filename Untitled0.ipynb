```python
#BETN1AI23050
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs

# 1. Generate Linearly Separable Data
X, y = make_blobs(n_samples=200, centers=2, random_state=42)
y = np.where(y == 0, -1, 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 2. Perceptron Implementation
class Perceptron:
    def __init__(self, lr=0.01, epochs=20):
        self.lr = lr
        self.epochs = epochs

    def fit(self, X, y):
        self.weights = np.zeros(X.shape[1])
        self.bias = 0

        for epoch in range(self.epochs):
            for i in range(len(X)):
                linear_output = np.dot(X[i], self.weights) + self.bias
                y_pred = 1 if linear_output >= 0 else -1
                update = self.lr * (y[i] - y_pred)
                self.weights += update * X[i]
                self.bias += update

            self.plot_decision_boundary(X, y, epoch)

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        return np.where(linear_output >= 0, 1, -1)

    def plot_decision_boundary(self, X, y, epoch):
        plt.scatter(X[:,0], X[:,1], c=y)
        x1 = np.linspace(min(X[:,0]), max(X[:,0]), 100)
        x2 = -(self.weights[0]*x1 + self.bias)/self.weights[1]
        plt.plot(x1, x2)
        plt.title(f"Epoch {epoch+1}")
        plt.show()

# 3. Train Model
model = Perceptron()
model.fit(X_train, y_train)

# 4. Accuracy
predictions = model.predict(X_test)
accuracy = np.mean(predictions == y_test)
print("Test Accuracy:", accuracy)
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

# 1. Load Data
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1,1)

encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Activation Functions
def relu(x): return np.maximum(0, x)
def relu_deriv(x): return x > 0
def softmax(x):
    exp = np.exp(x - np.max(x))
    return exp / exp.sum(axis=1, keepdims=True)

# Initialize Weights
np.random.seed(42)
W1 = np.random.randn(4, 10)
b1 = np.zeros((1,10))
W2 = np.random.randn(10, 3)
b2 = np.zeros((1,3))

lr = 0.01
epochs = 500

# Training
for epoch in range(epochs):
    # Forward
    z1 = X_train @ W1 + b1
    a1 = relu(z1)
    z2 = a1 @ W2 + b2
    a2 = softmax(z2)

    # Backprop
    dz2 = a2 - y_train
    dW2 = a1.T @ dz2
    db2 = np.sum(dz2, axis=0, keepdims=True)

    dz1 = dz2 @ W2.T * relu_deriv(z1)
    dW1 = X_train.T @ dz1
    db1 = np.sum(dz1, axis=0, keepdims=True)

    W2 -= lr * dW2
    b2 -= lr * db2
    W1 -= lr * dW1
    b1 -= lr * db1

# Evaluation
pred = np.argmax(softmax(relu(X_test @ W1 + b1) @ W2 + b2), axis=1)
true = np.argmax(y_test, axis=1)
print("Accuracy:", np.mean(pred == true))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

# Load MNIST small subset
mnist = fetch_openml('mnist_784')
X = mnist.data[:2000] / 255.0
y = mnist.target[:2000].astype(int)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

def sigmoid(x): return 1/(1+np.exp(-x))
def tanh(x): return np.tanh(x)
def relu(x): return np.maximum(0,x)

activations = {"ReLU": relu, "Sigmoid": sigmoid, "Tanh": tanh}

for name, activation in activations.items():
    W = np.random.randn(784, 64)
    V = np.random.randn(64, 10)
    lr = 0.01
    losses = []

    for epoch in range(10):
        hidden = activation(X_train @ W)
        output = hidden @ V
        loss = np.mean((output - y_train.reshape(-1,1))**2)
        losses.append(loss)

        V -= lr * hidden.T @ (output - y_train.reshape(-1,1))
        W -= lr * X_train.T @ ((output - y_train.reshape(-1,1)) @ V.T)

    plt.plot(losses, label=name)

plt.legend()
plt.title("Loss Comparison")
plt.show()
```

```python
def train_gd(X, y, batch_size=None):
    W = np.random.randn(X.shape[1],1)
    lr = 0.01
    losses = []

    for epoch in range(50):
        if batch_size:
            for i in range(0, len(X), batch_size):
                xb = X[i:i+batch_size]
                yb = y[i:i+batch_size]
                pred = xb @ W
                W -= lr * xb.T @ (pred - yb)
        else:
            pred = X @ W
            W -= lr * X.T @ (pred - y)

        loss = np.mean((X @ W - y)**2)
        losses.append(loss)
    return losses

X = np.random.randn(200,2)
y = np.random.randn(200,1)

loss_full = train_gd(X,y)
loss_mini = train_gd(X,y,batch_size=32)
loss_sgd = train_gd(X,y,batch_size=1)

plt.plot(loss_full,label="Full Batch")
plt.plot(loss_mini,label="Mini Batch")
plt.plot(loss_sgd,label="SGD")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Synthetic Data
X = np.random.randn(500, 4)
y = (np.sum(X, axis=1) > 0).astype(int).reshape(-1,1)

# Initialize Weights
W1 = np.random.randn(4,16)
W2 = np.random.randn(16,8)
W3 = np.random.randn(8,1)

lr = 0.01
losses = []

def sigmoid(x): return 1/(1+np.exp(-x))
def sigmoid_deriv(x): return sigmoid(x)*(1-sigmoid(x))

for epoch in range(200):
    # Forward
    z1 = X @ W1
    a1 = sigmoid(z1)
    z2 = a1 @ W2
    a2 = sigmoid(z2)
    z3 = a2 @ W3
    output = sigmoid(z3)

    loss = np.mean((output - y)**2)
    losses.append(loss)

    # Backprop
    d3 = (output - y) * sigmoid_deriv(z3)
    d2 = d3 @ W3.T * sigmoid_deriv(z2)
    d1 = d2 @ W2.T * sigmoid_deriv(z1)

    W3 -= lr * a2.T @ d3
    W2 -= lr * a1.T @ d2
    W1 -= lr * X.T @ d1

plt.plot(losses)
plt.title("Loss Curve")
plt.show()
```
