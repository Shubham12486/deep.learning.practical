```python
#BETN1AI23050

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Generate linearly separable data
np.random.seed(0)
X1 = np.random.randn(50,2) + [2,2]
X2 = np.random.randn(50,2) + [-2,-2]
X = np.vstack((X1,X2))
y = np.hstack((np.ones(50), np.zeros(50)))

# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

# Initialize weights
w = np.zeros(2)
b = 0
lr = 0.1
epochs = 10

def step(x):
    return 1 if x >= 0 else 0

# Training
for epoch in range(epochs):
    for i in range(len(X_train)):
        linear = np.dot(w, X_train[i]) + b
        y_pred = step(linear)
        error = y_train[i] - y_pred
        w += lr * error * X_train[i]
        b += lr * error

# Accuracy
correct = 0
for i in range(len(X_test)):
    if step(np.dot(w,X_test[i]) + b) == y_test[i]:
        correct += 1

print("Accuracy:", correct/len(X_test))
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

# Load data
iris = load_iris()
X = iris.data
y = iris.target.reshape(-1,1)

enc = OneHotEncoder()
y = enc.fit_transform(y).toarray()

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

# Initialize
np.random.seed(0)
W1 = np.random.randn(4,8)
b1 = np.zeros((1,8))
W2 = np.random.randn(8,3)
b2 = np.zeros((1,3))

def relu(x): return np.maximum(0,x)
def softmax(x): return np.exp(x)/np.sum(np.exp(x),axis=1,keepdims=True)

lr = 0.01
epochs = 200

for epoch in range(epochs):
    # Forward
    Z1 = np.dot(X_train,W1)+b1
    A1 = relu(Z1)
    Z2 = np.dot(A1,W2)+b2
    A2 = softmax(Z2)

    # Backprop
    dZ2 = A2 - y_train
    dW2 = np.dot(A1.T,dZ2)
    db2 = np.sum(dZ2,axis=0,keepdims=True)

    dA1 = np.dot(dZ2,W2.T)
    dZ1 = dA1*(Z1>0)
    dW1 = np.dot(X_train.T,dZ1)
    db1 = np.sum(dZ1,axis=0,keepdims=True)

    # Update
    W2 -= lr*dW2
    b2 -= lr*db2
    W1 -= lr*dW1
    b1 -= lr*db1

# Accuracy
pred = np.argmax(softmax(np.dot(relu(np.dot(X_test,W1)+b1),W2)+b2),axis=1)
true = np.argmax(y_test,axis=1)
print("Accuracy:", np.mean(pred==true))
```

```python
import matplotlib.pyplot as plt

def sigmoid(x): return 1/(1+np.exp(-x))
def tanh(x): return np.tanh(x)

activations = {
    "ReLU": relu,
    "Sigmoid": sigmoid,
    "Tanh": tanh
}

for name, func in activations.items():
    losses = []
    W = np.random.randn(4,8)
    for epoch in range(100):
        A = func(np.dot(X_train,W))
        loss = np.mean(A)
        losses.append(loss)
    plt.plot(losses,label=name)

plt.legend()
plt.title("Loss Comparison")
plt.show()
```

```python
 def train(batch_size):
    W = np.random.randn(4,3)
    losses=[]
    for epoch in range(100):
        for i in range(0,len(X_train),batch_size):
            Xb = X_train[i:i+batch_size]
            yb = y_train[i:i+batch_size]
            pred = np.dot(Xb,W)
            loss = np.mean((pred-yb)**2)
            losses.append(loss)
            grad = np.dot(Xb.T,(pred-yb))
            W -= 0.001*grad
    return losses

batch = train(len(X_train))
sgd = train(1)
mini = train(16)

plt.plot(batch,label="Batch")
plt.plot(sgd,label="SGD")
plt.plot(mini,label="Mini-batch")
plt.legend()
plt.show()
```

```python
np.random.seed(0)

W1 = np.random.randn(4,16)
W2 = np.random.randn(16,8)
W3 = np.random.randn(8,3)

lr = 0.001
epochs = 200
losses=[]

for epoch in range(epochs):
    # Forward
    A1 = relu(np.dot(X_train,W1))
    A2 = relu(np.dot(A1,W2))
    A3 = softmax(np.dot(A2,W3))

    loss = np.mean((A3-y_train)**2)
    losses.append(loss)

    # Backprop
    d3 = A3-y_train
    dW3 = np.dot(A2.T,d3)

    d2 = np.dot(d3,W3.T)*(A2>0)
    dW2 = np.dot(A1.T,d2)

    d1 = np.dot(d2,W2.T)*(A1>0)
    dW1 = np.dot(X_train.T,d1)

    W3 -= lr*dW3
    W2 -= lr*dW2
    W1 -= lr*dW1

plt.plot(losses)
plt.title("Loss Curve")
plt.show()
```
